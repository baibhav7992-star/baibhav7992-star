Perform:

```python
import pandas as pd
import numpy as np

# Given dataset
data = {
    'Age': [25, np.nan, 30, 35, np.nan],
    'Salary': [50000, 60000, np.nan, 80000, 75000]
}

df = pd.DataFrame(data)

# Print dataset
print("Dataset:")
print(df)

# Identify missing values
print("\nMissing Values:")
print(df.isnull())
```

• Fill missing values using mean


```python
import pandas as pd
import numpy as np

# Given dataset
data = {
    'Age': [25, np.nan, 30, 35, np.nan],
    'Salary': [50000, 60000, np.nan, 80000, 75000]
}

df = pd.DataFrame(data)

# Fill missing values using mean
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())

# Print updated dataset
print("Data after filling missing values:")
print(df)
```

Task 2: Outlier Detection (1 Mark)
Given:
df = pd.DataFrame({
 'Salary': [50000, 52000, 58000, 60000, 120000]
})

Perform:
• Plot boxplot


```python
import pandas as pd

df = pd.DataFrame({
    'Salary': [50000, 52000, 58000, 60000, 120000]
})

print(df)
```

• Identify outlier

```python
import pandas as pd

df = pd.DataFrame({
    'Salary': [50000, 52000, 58000, 60000, 120000]
})

# Calculate Q1, Q3 and IQR
Q1 = df['Salary'].quantile(0.25)
Q3 = df['Salary'].quantile(0.75)
IQR = Q3 - Q1

# Limits
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

# Identify outlier
outlier = df[(df['Salary'] < lower_limit) | (df['Salary'] > upper_limit)]

print("Outlier:")
print(outlier)
```

• Remove outlier using IQR

```python
import pandas as pd

df = pd.DataFrame({
    'Salary': [50000, 52000, 58000, 60000, 120000]
})

# Calculate Q1, Q3 and IQR
Q1 = df['Salary'].quantile(0.25)
Q3 = df['Salary'].quantile(0.75)
IQR = Q3 - Q1

# Limits
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

# Remove outlier
df_clean = df[(df['Salary'] >= lower_limit) & (df['Salary'] <= upper_limit)]

print("Data after removing outlier:")
print(df_clean)
```

Identify the mistake

• Rewrite correct code


```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Sample data
df = pd.DataFrame({
    'Age': [25, 30, 35, 40, 45],
    'Salary': [50000, 60000, 70000, 80000, 90000]
})

# Define X
X = df[['Age', 'Salary']]

# Split the data first
X_train, X_test = train_test_split(X, test_size=0.2, random_state=42)

# Create scaler
scaler = StandardScaler()

# Fit only on training data
X_train_scaled = scaler.fit_transform(X_train)

# Transform test data
X_test_scaled = scaler.transform(X_test)

# Print results
print("Training Data after Scaling:")
print(X_train_scaled)

print("\nTesting Data after Scaling:")
print(X_test_scaled)
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Dataset
data = {
    'Age': [22, 25, 47, 52, 46],
    'Income': [15000, 18000, 65000, 72000, 69000]
}

df = pd.DataFrame(data)

# Create scaler
scaler = MinMaxScaler()

# Normalize Income
df['Income_Normalized'] = scaler.fit_transform(df[['Income']])

print(df)
```

• Detect any anomalies


```python
import pandas as pd

# Dataset
data = {
    'Age': [22, 25, 47, 52, 46],
    'Income': [15000, 18000, 65000, 72000, 69000]
}

df = pd.DataFrame(data)

# IQR for Income
Q1 = df['Income'].quantile(0.25)
Q3 = df['Income'].quantile(0.75)
IQR = Q3 - Q1

# Limits
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

# Detect anomalies
anomalies = df[(df['Income'] < lower_limit) | (df['Income'] > upper_limit)]

print("Anomalies:")
print(anomalies)
```
