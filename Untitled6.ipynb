## Dataset Used
The following dataset is used for creating visualizations.

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample dataset
data = {
    'Age': [22, 25, 47, 52, 46, 56, 23, 34, 42, 29],
    'Income': [15000, 18000, 65000, 72000, 69000, 80000, 20000, 40000, 62000, 30000],
    'Experience': [1, 2, 20, 25, 18, 30, 1, 7, 15, 4],
    'Spending': [2000, 2500, 7000, 8500, 7800, 9000, 2200, 4500, 7200, 3500]
}

df = pd.DataFrame(data)

print("Dataset:")
print(df)
```

## Task 1: Create Histogram

### Question:
Create a histogram for the **Income** column to understand its distribution.

```python
plt.figure(figsize=(6,4))
plt.hist(df['Income'], bins=5, edgecolor='black')
plt.title("Histogram of Income")
plt.xlabel("Income")
plt.ylabel("Frequency")
plt.show()
```

### Interpretation:
The histogram shows the distribution of income values.  
Most individuals are in the medium to high income range, while fewer people are in the low income range.

## Task 2: Create Boxplot

### Question:
Create a boxplot for the **Income** column to detect spread and outliers.

```python
plt.figure(figsize=(6,4))
plt.boxplot(df['Income'])
plt.title("Boxplot of Income")
plt.ylabel("Income")
plt.show()
```

### Interpretation:
The heatmap shows the correlation between Age, Income, Experience, and Spending.  
A positive correlation means that as one variable increases, the other also tends to increase.

## Task 4: Create Scatter Plot

### Question:
Create a scatter plot between **Age** and **Income** to identify their relationship.

```python
plt.figure(figsize=(6,4))
plt.scatter(df['Age'], df['Income'])
plt.title("Scatter Plot of Age vs Income")
plt.xlabel("Age")
plt.ylabel("Income")
plt.show()
```

### Interpretation:
The scatter plot shows a positive relationship between age and income.  
As age increases, income generally tends to increase.

## Task 5: Create Bar Chart

### Question:
Create a bar chart for the **Spending** column to compare spending values among individuals.

```python
plt.figure(figsize=(8,4))
plt.bar(df.index, df['Spending'])
plt.title("Bar Chart of Spending")
plt.xlabel("Person Index")
plt.ylabel("Spending")
plt.show()
```

### Interpretation:
The bar chart compares spending values of different individuals.  
It clearly shows which individuals have higher or lower spending.

## Task 6: Interpret Patterns and Insights from Visualizations

### Final Insights:
1. Income is distributed from low to high values.

3. The heatmap shows strong positive relationships between variables.
4. Age and income are positively related.
5. Spending tends to increase with higher income.
