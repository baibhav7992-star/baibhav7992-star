```python
# Case Study 1: Student Marks Analysis
import numpy as np

# Student marks dataset
marks = np.array([
    [80, 75, 90],
    [60, 88, 70],
    [95, 92, 85],
    [70, 65, 60],
    [88, 78, 84]
])

print("Marks Dataset:\n", marks)

# 1. Total marks of each student
total_marks = np.sum(marks, axis=1)
print("\nTotal Marks of Each Student:", total_marks)

# 2. Average marks per student
average_marks = np.mean(marks, axis=1)
print("\nAverage Marks per Student:", average_marks)

# 3. Identify topper
topper_index = np.argmax(total_marks)
print("\nTopper is Student:", topper_index + 1)

# 4. Subject-wise highest score
highest_subject = np.max(marks, axis=0)
print("\nHighest Score in Each Subject:", highest_subject)

# 5. Standard deviation of each subject
std_subject = np.std(marks, axis=0)
print("\nStandard Deviation of Each Subject:", std_subject)
```

```python
import numpy as np

# Student marks dataset
marks = np.array([
    [80, 75, 90],
    [60, 88, 70],
    [95, 92, 85],
    [70, 65, 60],
    [88, 78, 84]
])

print("Marks Dataset:\n", marks)

# 1. Total marks of each student
total_marks = np.sum(marks, axis=1)
print("\nTotal Marks of Each Student:", total_marks)

# 2. Average marks per student
average_marks = np.mean(marks, axis=1)
print("\nAverage Marks per Student:", average_marks)

# 3. Identify topper
topper_index = np.argmax(total_marks)
print("\nTopper is Student:", topper_index + 1)

# 4. Subject-wise highest score
highest_subject = np.max(marks, axis=0)
print("\nHighest Score in Each Subject:", highest_subject)

# 5. Standard deviation of each subject
std_subject = np.std(marks, axis=0)
print("\nStandard Deviation of Each Subject:", std_subject)
```

## Interpretation – Case Study 1

1. Student 3 has the highest total marks and is identified as the topper.
2. Subject-wise highest scores indicate strong academic performance.
3. Standard deviation shows variation in marks across subjects.
4. Students with lower deviation show more consistent performance.
5. NumPy vectorized operations make matrix computations efficient and simple.


```python
# Case Study 3: Sales Analysis
# Sales dataset
sales = np.array([1200, 1500, 1100, 1700, 1600, 1800, 1400])

print("Sales Data:", sales)

# 1. Average daily sales
average_sales = np.mean(sales)
print("\nAverage Daily Sales:", average_sales)

# 2. Day with highest sales
highest_day = np.argmax(sales)
print("\nHighest Sales on Day:", highest_day + 1)

# 3. Percentage growth from Day 1 to Day 7
growth = ((sales[-1] - sales[0]) / sales[0]) * 100
print("\nPercentage Growth from Day 1 to Day 7:", growth, "%")

# 4. Cumulative sales
cumulative_sales = np.cumsum(sales)
print("\nCumulative Sales:", cumulative_sales)

# 5. Standard deviation
std_sales = np.std(sales)
print("\nStandard Deviation of Sales:", std_sales)
```

## Interpretation – Case Study 3

1. The average daily sales provide an overview of weekly performance.
2. The highest sales were recorded on Day 6.
3. There is an overall positive growth from Day 1 to Day 7.
4. Cumulative sales show steady revenue increase.
5. Standard deviation indicates moderate fluctuation in sales.
